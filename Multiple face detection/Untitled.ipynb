```python
import numpy as np
from matplotlib import pyplot as plt
import cv2
import glob
```

```python
txtfiles = [] 
for file in glob.glob("man3.jpg"):
    txtfiles.append(file)
```

```python
for i in txtfiles:
    img = cv2.imread(i,cv2.IMREAD_COLOR)
    img = img.copy()
    #img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import matplotlib.image as mpimg
img = mpimg.imread('man3.jpg')
imgplot = plt.imshow(img)
plt.show()
```

```python
facecascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
```

```python
faces = facecascade.detectMultiScale(img, scaleFactor=1.2, minNeighbors=5)
```

```python
print('Total number of Faces found',len(faces))
```

```python
for (x, y, w, h) in faces:
    face_detect = cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 255), 2)
    roi_gray = img[y:y+h, x:x+w]
    roi_color = img[y:y+h, x:x+w]       
    plt.imshow(face_detect) 
    plt.show()
```

```python
eyes = eye_cascade.detectMultiScale(roi_gray)
for (ex,ey,ew,eh) in eyes:
    eye_detect = cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(255,0,255),2)
    plt.imshow(eye_detect)
    plt.show()
```
